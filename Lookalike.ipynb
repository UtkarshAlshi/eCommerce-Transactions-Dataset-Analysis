```python
customer_features = data.groupby('CustomerID').agg({
    'TotalValue': 'sum',  # Total transaction value per customer
    'Quantity': 'sum',    # Total quantity of products purchased
    'Price_x': 'mean'       # Average price of products purchased
}).reset_index()

print(customer_features.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(customer_features[['TotalValue', 'Quantity', 'Price_x']])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(features_scaled)
```

```python
similarity_df = pd.DataFrame(similarity_matrix,
                             index=customer_features['CustomerID'],
                             columns=customer_features['CustomerID'])

print(similarity_df.head())
```

```python
lookalikes = {}

for cust_id in customer_features['CustomerID'][:20]:
    # Sort the similarity scores for the current customer in descending order
    similar_customers = similarity_df[cust_id].sort_values(ascending=False).iloc[1:4]
    # Store the top 3 similar customers and their similarity scores
    lookalikes[cust_id] = list(zip(similar_customers.index, similar_customers.values))

print(lookalikes)
```

```python
import csv

output_path = '/content/drive/My Drive/eCommerce_Assignment/Utkarsh_Alshi_Lookalike.csv'

with open(output_path, mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['CustomerID', 'Lookalikes'])

    for cust_id, sim_list in lookalikes.items():
        writer.writerow([cust_id, sim_list])

print(f"Lookalike data saved to {output_path}")
```
